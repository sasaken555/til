### データの再利用

一般的なデータの読み書きだけでなく、Pythonのオブジェクトをそのまま読み書きして再利用することもできる。  
ここではPython標準ライブラリのpickleを使う。


```python
df_generic_domains = tables[1]
df_generic_domains.to_pickle("generic_top_domains_wiki.pickle")
!ls -l generic_top_domains_wiki.pickle
```

```python
df_from_pickle = pd.read_pickle("generic_top_domains_wiki.pickle")
df_from_pickle.head() 
```
