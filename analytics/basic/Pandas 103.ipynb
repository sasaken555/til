### データ型変換

```python
df.dtypes
```

```python
# 日付のデータ型を文字列 --> datetimeに変換
df.loc[:, "date"] = df.loc[:, "日付"].apply(pd.to_datetime)
df.loc[:, "date"]
```

```python
df.head(10)
```

```python
# in64 --> float32
df.loc[: , "摂取カロリー"] = df.loc[: , "摂取カロリー"].astype(np.float32)

# dateカラムをindexにセット
df = df.set_index("date")
df.head()
```

### 並べ替え

```python
df.sort_values(by="歩数")
```

```python
# sort_valuesはデフォルトで昇順ソートになる。降順ソートを実施してみる。
df.sort_values(by="歩数", ascending=False).head()
```

### 不要カラムの削除 

```python
# dateがindexになったので、日付カラムを落とす
df = df.drop("日付", axis=1)
df.tail()
```

### 組み合わせデータの挿入

```python
df.loc[:, "歩数/カロリー"] = df.loc[: , "歩数"] / df.loc[: , "摂取カロリー"]
df.head()
```

```python
# 関数を適用して新しい列を追加してみる
def exercise_judge(ex):
    if ex <= 3.0:
        return "Low"
    elif ex <= 6.0:
        return "Mid"
    else:
        return "High"

df.loc[: , "運動指数"] = df.loc[: , "歩数/カロリー"].apply(exercise_judge)
df.head(10)
```

```python
# ここまでの結果を保存しよう
df.to_pickle("201704_health_calc.pickle")
```
