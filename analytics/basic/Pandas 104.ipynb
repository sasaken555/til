pandasで時系列データを操作します。

```python
# 1ヵ月分のデータを作る
import pandas as pd
dates = pd.date_range(start="2017-04-01", end="2017-04-30")
dates
```

```python
import numpy as np
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1, 31, 30), index=dates, columns=["乱数"])
df
```

```python
# 1年分 (=365日)のデータを作る
yearly_dates = pd.date_range(start="2017-01-01", periods=365)
yearly_dates
```

```python
# 1年分のデータでDataFrameを作成
np.random.seed(123)
df_yearly = pd.DataFrame(np.random.randint(1, 31, 365), index=yearly_dates, columns=["乱数"])
df_yearly.head(10)
```

```python
# 月平均のデータ
df_yearly.groupby(pd.Grouper(freq="M")).mean()
```

```python
# df.resample() でもグルーピングできるので、ここから平均値を取れる
# カラムを固定すれば戻り値はSeries型
df_yearly.loc[ : , "乱数" ].resample("M").mean()
```

```python
# 土曜日区切りの1週間ごとの合計値のDataFrame
df_yearly_by_saturday = pd.DataFrame(df_yearly.groupby(pd.Grouper(freq="W-SAT")).sum(), columns=["乱数"])
df_yearly_by_saturday
```
