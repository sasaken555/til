```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
```

## MATLABスタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画
plt.plot(x, y)  # 折れ線グラフ
plt.title('MATLAB-style')  # グラフタイトルの指定
plt.show()  # グラフ表示
```

## オブジェクト指向スタイル

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクトとサブプロットを生成
fig, ax = plt.subplots()

ax.plot(x, y)  # 折れ線グラフ
ax.set_title('OOP-style')  # グラフタイトルの指定

plt.show()  # グラフ表示
```

```python
# 1つの描画オブジェクトに対して複数のサブプロットを載せる
fig, axes = plt.subplots(2)  # 2つのサブプロット
plt.show()
```

```python
# 行ど列数も指定可能
fig, axes = plt.subplots(2, 2)  # 2行2列のサブプロット
plt.show()
```

```python
# plt.subplotsの引数はnrows, ncolsで行と列を指定できる
# デフォルト値はそれぞれ1
# 列だけ指定は以下の通り
fig, axes = plt.subplots(ncols=2)  # 1行2列のサブプロット
plt.show()
```

```python
# グラフスタイルは26種類用意されている。好きなスタイルを選ぼう！
print(matplotlib.style.available)
```

```python
# classic スタイル
matplotlib.style.use('classic')
fig, ax = plt.subplots()
ax.plot([1, 2])
plt.show()
```

```python
# 描画オブジェクトや各サブプロットに別々のタイトルをつける
matplotlib.style.use('ggplot')
fig, axes = plt.subplots(ncols=2)

# サブプロット
axes[0].set_title('subplot title 0')
axes[1].set_title('subplot title 1')

# 描画オブジェクト
fig.suptitle('figure title')

plt.show()
```

```python
# 軸ラベル
fig, ax = plt.subplots()
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

```python
# 凡例
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 5, 9], label='legend label')
ax.legend(loc='best')  # 凡例を表示
plt.show()
```

### グラフの種類と出力方法

```python
# 2つの折れ線グラフの描画
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]

ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```

```python
# 線形ならば軸の値間隔を狭めて曲線の描画もできる
import numpy as np

x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```

```python
# 棒グラフ
fig, ax = plt.subplots()
x = [1, 2, 3]
y = [10, 2, 3]
labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label=labels)
plt.show()
```

```python
# X軸をずらせば複数の棒グラフを載せることができる
# 棒グラフ
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

width = 0.4
ax.bar(x, y1, width=width, tick_label=labels, label='y1')

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='y2')

ax.legend()
plt.show()
```

```python
# 積み上げ棒グラフ
fig, ax = plt.subplots()
x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]
ax.bar(x, y_total, tick_label=labels, label='y1')
ax.bar(x, y2,  label='y2')

ax.legend()
plt.show()
```

```python
# 散布図
fig, ax = plt.subplots()
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y)
plt.show()
```

```python
# ヒストグラム
# データ生成
mu = 100  # 平均値
sigma = 15  # 標準偏差
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()
n, bins, pathced = ax.hist(x)
plt.show()
```

```python
# 箱ひげ図
matplotlib.style.use('seaborn')

x0 = np.random.normal(0, 10, 500)
x1 = np.random.normal(0, 15, 500)
x2 = np.random.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']
ax.boxplot((x0, x1, x2), labels=labels)

plt.show()
```

```python
# 円グラフ
x = [10, 3, 5]
labels = ['spam', 'ham', 'egg']

fig, ax = plt.subplots()
ax.pie(x, labels=labels, startangle=90, counterclock=False, autopct='%1.2f%%')
ax.axis('equal')  # アスペクト比を保持する。指定しない場合は領域に合わせて楕円になることが多い。。。

plt.show()
```

### 複数のグラフを組み合わせる

```python
fig, ax = plt.subplots()
x1 = [1, 2, 3]
y1 = [5, 2, 3]
x2 = [1, 2, 3, 4]
y2 = [8, 5, 4, 6]

ax.bar(x1, y1, label='y1')
ax.plot(x2, y2, label='y2')
ax.legend()

plt.show()
```

### Pandasオブジェクトからグラフ描画

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [3, 1, 2]})
df.plot()  # 折れ線グラフの描画。内部的にはMatplotlibを使用している。
plt.show()
```

```python
np.random.seed(123)
df2 = pd.DataFrame(np.random.rand(3, 2), columns=['y1', 'y2'])
df2.plot.bar()  # 棒グラフを描画
plt.show()
```
