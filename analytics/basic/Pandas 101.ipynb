## Series

## DataFrame

```python
# 2次元データ DataFrame
df = pd.DataFrame([[10, "a", True],
                  [20, "b", False],
                  [30, "c", False],
                  [40, "d", True]])

df
```

```python
# numpyの
import numpy as np
df = pd.DataFrame(np.arange(100).reshape((25, 4)))
df
```

```python
# 最初・最後からN行を表示
print("------------ df.head() ------------")
print(df.head())
print("------------ df.tail() ------------")
print(df.tail())
```

```python
# DataFrameの型を知る
df.shape
```

```python
# Index名、Column名は自分でつけることができる
df = pd.DataFrame(np.arange(6).reshape((3, 2)))
df.index= ["01", "02", "03"]
df.columns = ["A", "B"]
df
```

```python
# DataFrame生成時にIndex/Column名をつけることもできる
df = pd.DataFrame(np.arange(8).reshape((4, 2)),
                 columns=["A列", "B列"],
                 index=["1行目", "2行目", "3行目", "4行目"])
df
```

```python
# dict型のデータからDataFrameを作る
pd.DataFrame({"A列": [1, 2, 3], "B列": [6, 7, 8]})
```

## DataFrameからデータ抽出

```python
# データの抽出
df = pd.DataFrame(np.arange(12).reshape((4, 3)),
                 columns=["A", "B", "C"],
                 index=["1行目", "2行目", "3行目", "4行目"])
df
```

```python
# 単一カラムを指定. 1次元のデータなので、Series型になる
df["A"]
```

```python
# 複数カラムを指定して抽出
df[["A", "B"]]
```

```python
# もちろんIndexを指定して抽出もできr
df[:2]
```

```python
# 引数にlistを渡すのではなく、メソッドを使って抽出
df.loc[:, :]  # 全部
```

```python
# locメソッドは .loc[行, 列] のインデックスを指定
df.loc[:, "A"]
```

```python
# = df[["A", "B"]]
df.loc[:, ["A", "B"]]
```

```python
# Index方向に抽出
df.loc["3行目", :]
```

```python
# もちろんIndex方向の抽出で複数条件を指定することも可能
df.loc[["1行目", "2行目"], :]
```

```python
# Index/Column両方の条件を指定してDataFrameを抽出
df.loc[["4行目"], ["B", "C"]]
```

```python
# 位置のIndexを直接指定して抽出するときは、iloc メソッドを使う
df.iloc[1, 1]
```

```python
df.iloc[1:, 1]
```
